```python

#1
boat_side = 'Right'
missionaries_on_right = 3
cannibals_on_right = 3
missionaries_on_left = 0
cannibals_on_left = 0

#2
print('\U0001f482' * missionaries_on_left,  '\U0001f479' *cannibals_on_left,'\U0001f6A2''\U0001f30a'*5 ,'\U0001f6A2','\U0001f482 '* missionaries_on_right, '\U0001f479'*cannibals_on_right)

#3
while True:
  missionaries = int(input('Enter the number of missionaries : '))
  cannibals = int(input('Enter the number of cannibals : '))

  #4
  if missionaries + cannibals < 1 or missionaries + cannibals > 2:
    print('invalid Move 1')
    continue
  else:
    #5
    if boat_side == 'Right':
      print('The boat is on right side')


      #6
      if missionaries > missionaries_on_right or cannibals > cannibals_on_right:
        print('invalid move 2')
        continue
      else:
        print('valid move2')


        #7
        boat_side = 'Left'
        missionaries_on_right=missionaries_on_right-missionaries
        cannibals_on_right=cannibals_on_right-cannibals
        missionaries_on_left=missionaries_on_left+missionaries
        cannibals_on_left=cannibals_on_left+cannibals

        #8
        print('\U0001f482'* missionaries_on_left, '\U0001f479'* cannibals_on_left, '\U0001f30a'*5, '\U0001f482'* missionaries_on_right, '\U0001f479'*cannibals_on_right)

    #9
    elif boat_side=="Left":
      print('the boat is on Left')


      #10
      if missionaries > missionaries_on_left or cannibals > cannibals_on_left:
        print("invalid move 2")
        continue
      else:
        print("valid move 2")
        #11
        boat_side="Right"
        missionaries_on_right=missionaries_on_right+missionaries
        cannibals_on_right=cannibals_on_right+cannibals
        missionaries_on_left=missionaries_on_left-missionaries
        cannibals_on_left=cannibals_on_left-cannibals




        #12
        print('\U0001f482'* missionaries_on_left, '\U0001f479'* cannibals_on_left, '\U0001f30a'*5, '\U0001f482'* missionaries_on_right, '\U0001f479'*cannibals_on_right)



    if missionaries_on_right<cannibals_on_right and missionaries_on_right>0:
      print('YOU LOSE')
      break
    elif missionaries_on_left<cannibals_on_left and missionaries_on_left>0:
      print('YOU LOSE')
      break


    if missionaries_on_left==3 and cannibals_on_left==3:
      print('YOU WIN')
      break

print("Game Over")
```
